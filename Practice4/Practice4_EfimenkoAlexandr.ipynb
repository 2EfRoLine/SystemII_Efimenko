#### Задание 2. Произвести предобработку данных dataset "Титаник" и сохранить результаты в каждый отдельный датафрейм
1) Удалить все пустые значения и сохранить результаты в датафрейм df_del
2) Заменить на конкретное значение все пустые и сохранить результаты в датафрейм df_num
